```python
# check for missing values
df.isnull().sum()
```

```python
# missing some values for column Review Rating, replace null values with median
# of its relevant category's review rating
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
df.isnull().sum()
```

```python
# rename columns to snake case for better standalization
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df.rename(columns={'purchase_amount_(usd)':'purchase_amount'}, inplace = 'True')
```

```python
# create a column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
df['frequency_of_purchases'].unique()
```

```python
# create column purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly':14,
    'Annually': 365,
    'Every 3 Months': 90   
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
